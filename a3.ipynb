```python
import seaborn as sn
```

```python
df = sn.load_dataset('iris')
```

```python
import numpy as np
```

```python
df.groupby('species').agg(['median', 'mean', 'min', 'max', 'std'])
```

```python
df.groupby('species')['sepal_length'].agg(['median', 'mean', 'min', 'max', 'std'])
```

```python
df.groupby('species')['sepal_length'].apply(list)
```

```python
species = df['species'].unique()
for name in species:
    print(f'Species: {name}')
    print(df[df['species'] == name].describe())
    print()
```
