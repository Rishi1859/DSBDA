```python
dataset = sn.load_dataset('iris')
```

```python
dataset.dtypes
```

```python
sn.histplot(dataset['petal_width'])
```

```python
sn.histplot(dataset['petal_length'])
```

```python
sn.histplot(dataset['sepal_width'])
```

```python
sn.histplot(dataset['sepal_length'])
```

```python
# from the histogram we can observe that 
# the distribution of petal length and petal 
# width is left skewed.
```

```python
import matplotlib.pyplot as plt

_ , axes = plt.subplots(2,2)
plt.subplots_adjust(wspace=0.3, hspace=0.2)
sn.boxplot(dataset['petal_length'], ax=axes[0][0])
sn.boxplot(dataset['sepal_length'], ax=axes[0][1])
sn.boxplot(dataset['petal_width'], ax=axes[1][0])
sn.boxplot(dataset['sepal_width'], ax=axes[1][1])
```
