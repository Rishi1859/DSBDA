```python
import seaborn as sn
```

```python
dataset = sn.load_dataset('titanic')
```

```python
dataset.head()
```

```python
sn.scatterplot(x=dataset['age'], y=dataset['fare'], hue=dataset['sex'])
```

```python
# observation: the fare for the male and 
# female seems uniform. except for two male datapoints
```

```python
sn.histplot(dataset['fare'])
```

```python
# seems like most of the people have paid 0 for the fare.
```

```python
sn.boxplot(x=dataset['sex'], y=dataset['age'], hue=dataset['survived'])
```

```python
# from the box plot we can conclude that mostly younger males
# and older females have survived.
```

```python
sn.barplot(x='sex', y='survived', data=dataset)
```

```python
# more females have survived as compared to male
```
